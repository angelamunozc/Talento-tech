### Matriz de Correlacin

Una matriz de correlacin muestra la correlacin entre diferentes variables numricas en un DataFrame. Los valores varan entre -1 y 1, donde:
- 1 indica una correlacin positiva perfecta.
- -1 indica una correlacin negativa perfecta.
- 0 indica que no hay correlacin.

Podemos usar la matriz de correlacin para ver cmo se relacionan el peso, la altura y el IMC en nuestro DataFrame `df`.

```python
# Seleccionar solo las columnas numricas relevantes
df_numeric = df[["Peso (kg)", "Altura (m)", "IMC"]]

# Calcular la matriz de correlacin
correlation_matrix = df_numeric.corr()

# Mostrar la matriz de correlacin
print("Matriz de Correlacin:")
display(correlation_matrix)
```

### Matriz de Confusin

Una matriz de confusin es una tabla que se utiliza para evaluar el rendimiento de un modelo de clasificacin. Muestra el nmero de predicciones correctas e incorrectas para cada clase.

Para este ejemplo, generar algunos datos de ejemplo para una tarea de clasificacin simple, ya que nuestro DataFrame actual no es adecuado para generar una matriz de confusin real de un modelo.

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos de ejemplo (valores reales y predicciones)
# Supongamos un problema de clasificacin con dos clases: "A" y "B"
y_true = ["A", "B", "A", "A", "B", "B", "A", "B", "A", "A"]
y_pred = ["A", "A", "A", "B", "B", "B", "A", "A", "A", "B"]

# Calcular la matriz de confusin
cm = confusion_matrix(y_true, y_pred)

# Mostrar la matriz de confusin
print("\nMatriz de Confusin:")
display(cm)

# Visualizar la matriz de confusin (opcional, pero recomendado)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["A", "B"], yticklabels=["A", "B"])
plt.xlabel("Prediccin")
plt.ylabel("Valor Real")
plt.title("Matriz de Confusin de Ejemplo")
plt.show()
```
