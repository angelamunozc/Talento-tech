### Matriz de Correlación

Una matriz de correlación muestra la correlación entre diferentes variables numéricas en un DataFrame. Los valores varían entre -1 y 1, donde:
- 1 indica una correlación positiva perfecta.
- -1 indica una correlación negativa perfecta.
- 0 indica que no hay correlación.

Podemos usar la matriz de correlación para ver cómo se relacionan el peso, la altura y el IMC en nuestro DataFrame `df`.

```python
# Seleccionar solo las columnas numéricas relevantes
df_numeric = df[["Peso (kg)", "Altura (m)", "IMC"]]

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Mostrar la matriz de correlación
print("Matriz de Correlación:")
display(correlation_matrix)
```

### Matriz de Confusión

Una matriz de confusión es una tabla que se utiliza para evaluar el rendimiento de un modelo de clasificación. Muestra el número de predicciones correctas e incorrectas para cada clase.

Para este ejemplo, generaré algunos datos de ejemplo para una tarea de clasificación simple, ya que nuestro DataFrame actual no es adecuado para generar una matriz de confusión real de un modelo.

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos de ejemplo (valores reales y predicciones)
# Supongamos un problema de clasificación con dos clases: "A" y "B"
y_true = ["A", "B", "A", "A", "B", "B", "A", "B", "A", "A"]
y_pred = ["A", "A", "A", "B", "B", "B", "A", "A", "A", "B"]

# Calcular la matriz de confusión
cm = confusion_matrix(y_true, y_pred)

# Mostrar la matriz de confusión
print("\nMatriz de Confusión:")
display(cm)

# Visualizar la matriz de confusión (opcional, pero recomendado)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["A", "B"], yticklabels=["A", "B"])
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.title("Matriz de Confusión de Ejemplo")
plt.show()
```
